```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# 準備數據...  假設我要畫一個 sin波 從 0~180度
# 設定要畫的 x,y數據 list
x = np.arange(0, 180)
y = np.cos(x * np.pi/180.0)

# 開始畫圖
plt.plot(x,y)
# 在這個指令之前，都還在做畫圖的動作
# 這個指令算是 "秀圖"
plt.show()
```

```python
# 畫出完整的 sin 圖形
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)

plt.plot(x, y_sin)

plt.show()
```

```python
# 畫出完整的 cos 圖形
x = np.arange(0, 2 * np.pi, 0.1)
y = np.cos(x)

plt.plot(x, y)

plt.savefig("D18_cos.png", dpi = 300, format = "png")
```

```python
n = 1024
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)
r = np.arctan2(Y, X)

plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)

plt.scatter(X, Y, s = 50, c=r, alpha = 0.5)
plt.show()
```

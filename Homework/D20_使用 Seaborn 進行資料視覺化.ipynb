```python
import numpy as np
import seaborn as sns
```

```python
# 設定圖形樣式 - whitegrid
# use sns.set
sns.set_style("whitegrid")

# 利用 NUMPY去建立資料集
# np.random.Randomstate 設定數學式
rs = np.random.RandomState()
x = rs.normal(2, 1, 75)
y = 2 + 1.5*x + rs.normal(0, 2, 75)

# 畫圖
# sns.residplot
sns.residplot(x, y)
```

```python
# bin：指的是特徵值
x1 = np.random.normal(size = 100)
sns.distplot(x1)
# kde：on / off
# sns.distplot()
```

```python
import matplotlib.pyplot as plt

f = plt.figure()
P1 = f.add_subplot(2, 2, 1)
sns.distplot(x1, bins = 20, kde = False)
P2 = f.add_subplot(2, 2, 2)
sns.distplot(x1, bins = 20)
P3 = f.add_subplot(2, 2, 3)
sns.distplot(x1, bins = 5, kde = False)
P4 = f.add_subplot(2, 2, 4)
sns.distplot(x1, bins = 5)

plt.show()
```

```python
# 1. bin的大小決定數據的精度與分布的範圍，bin值越大數據就越接近實際值，卻也使得分布的圖形廣度增大
# 2. 有無 kde對於資料的圖形與分布沒有影響，但是改變資料代表的意義，從實數轉換成機率的分布
```

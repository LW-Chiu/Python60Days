```python
# 載入套件
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 創建基本地圖
# 當 projection參數為 'ortho'，所得圖為地球儀截面

# 設定圖形大小
plt.figure(figsize = (5, 5))

# 設定投影型態
m = Basemap(projection = 'ortho', lat_0 = 50, lon_0 = -100)

# 設定圖形顏色，選用 NASA寶石藍輸出
m.bluemarble()

# 秀圖
plt.show()
```

```python
# 設置 Lambert保形底圖
# width = 12000000, height = 9000000
# 經緯度 lat_1 = 45., lat_2 = 55, lat_0 =50, lon_0 = -107
# set resolution = None 跳過邊界數據集的處理
# projection = 'lcc'

m = Basemap(projection = 'lcc', lat_1 = 45, lat_2 = 55, lat_0 = 50, lon_0 = -107, width = 12000000, height = 9000000, resolution = None)

# 選用 NASA寶石藍輸出
m.bluemarble()

# 秀圖
plt.show()
```

```python
# projection 參數為 'Mill'
# 設定經緯座標
# 左下角的緯度 - llcrnrlat = -90
# 左下角的經度 - llcrnrlon = -180
# 右上角的緯度 - urcrnrlat = 90
# 右下角的經度 - urcrnrlon = 180

# 成圖設定 resolution

m = Basemap(projection = 'mill', llcrnrlat = -90, llcrnrlon = -180, urcrnrlat = 90, urcrnrlon = 180, resolution = 'l')

# 創建邊線
m.drawcoastlines()
# 畫出國家，使用線寬為 2的線條生成分界線
m.drawcountries(linewidth = 2)
# 畫出州界
m.drawstates(color = 'b')
# 畫出城市
m.drawcounties(color = 'darkred')

plt.title('Basemap Tutorial')
plt.show()
```
